````python
# クリップボード操作するライブラリー
import pyperclip
````

````python
code = ""
code += "func main() { \n"
for i in range(3):
    code += "    ch{} := make(chan bool)\n".format(i)
for i in range(3):
    code += "    go func() { \n"
    code += "        request() \n"
    code += "        ch{} <- true \n".format(i)
    code += "    }()\n"
for i in range(3):
    code += "    <-ch{}\n".format(i)
code += "}"

print(code)
pyperclip.copy(code)
````
